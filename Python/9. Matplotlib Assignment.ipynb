We have the min and max temperature in a city in india for each month of the year. 
we would like to find a function to describe this and show it graphically, the dataset given below. 
Task:- 
    1. fitting it to the periodic function 
    2. plot the fit 
Data:- 

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25  <br> 
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, 'go')
plt.plot(months, temp_min, 'co')
plt.xlabel('Month')
plt.ylabel('Min and max temperature')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

print(max_temp)
print(min_temp)
```

```python
months = np.arange(12)
months
```

```python
plt.xlabel("Months")
plt.ylabel("Temperature in celcius")
plt.plot(months,max_temp,'rs',label = "max_temp")
plt.plot(months,min_temp,'bs',label = "min_temp")
plt.legend()
plt.grid()
plt.show()
```

**<font size="4">Fitting it into periodic function</font>**

```python
from scipy import optimize
def yearly_temp(time,avg,amplitude,timeOffset): 
    return (avg + amplitude * np.cos((time + timeOffset) * 1.8 * np.pi / time.max()))

res_Max, cov_max = optimize.curve_fit(yearly_temp,months,max_temp,[40,20,0])
res_Min, cov_min = optimize.curve_fit(yearly_temp,months,min_temp,[-40,20,0])
```

**<font size="4"> Lets plot the Fit</font>**

```python
days = np.linspace(0,12, num=365)
plt.figure()
plt.plot(months,max_temp,'rs')
plt.plot(days,yearly_temp(days,*res_Max),'g-')
plt.plot(months,min_temp,'bs')
plt.plot(days,yearly_temp(days,*res_Min),'y-')
plt.xlabel("Months")
plt.ylabel("Temperature in celcius")
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
```

```python
titanic_data = pd.read_csv(url) 
```

**<font size =4> 1. Create a pie chart presenting the male/female proportion</font>**

```python
s1 = round((titanic_data["sex"].value_counts())/len(titanic_data)*100,2) 
pd.DataFrame(s1)
```

```python
chart_labels =["Male","Female"] 
sizes = titanic_data.sex.value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(sizes, autopct="%1.1f%%", labels = chart_labels, colors = ["red","blue"])
plt.show()
```

**<font size =4> 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender</font>**

```python
plt.figure()
category1 = titanic_data[titanic_data.sex=="male"].plot.scatter("age","fare",color="blue",label="male")
titanic_data[titanic_data.sex == "female"].plot.scatter("age","fare",color="green",label="female",ax=category1)
plt.show()
```
