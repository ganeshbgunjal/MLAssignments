```python
import pandas as pd
```

```python
# to get iris dataset
site_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

iris = pd.read_csv(site_url, names= ["Sepal length", "Sepal width", "Petal length", "Petal width", "Species"])
```

```python
iris
```

```python
features = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
```

```python
x = iris.loc[:,features].values # got values of  SL,PL,SW,PW
x
```

```python
y = iris.loc[:,["Species"]].values # got types of iris flower
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
principalComponents = pca.fit_transform(x)
pricipalDf = pd.DataFrame(data=principalComponents,columns= ["Principal Component-1","Principal Component-2","Principal Component-3"])
```

```python
Final_dataframe = pd.concat([pricipalDf,iris[["Species"]]],axis = 1)  #column wise concatenation
```

```python
# For plotting 3d values we need values for x axis, y axis, z axis
x_values = Final_dataframe['Principal Component-1']
y_values = Final_dataframe['Principal Component-2']
z_values = Final_dataframe['Principal Component-3']
```

```python
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import random
#creating figure

fig = plt.figure(figsize =(10,10))  #layout of 10*10
ax = fig.add_subplot(111,projection = "3d")
ax.set_xlabel("Principal Component-1",fontsize  = 15) #label on x axis
ax.set_ylabel("Principal Component-2",fontsize  = 15) #label on y axis
ax.set_zlabel("Principal Component-3",fontsize  = 15) #label on z axis
ax.set_title("PCA Components", fontsize = 20)

Species_types = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = 'r','g','b'



for Species_type,color in zip(Species_types,colors): 
    Indices = Final_dataframe["Species"] == Species_type
    ax.scatter(Final_dataframe.loc[Indices, 'Principal Component-1'],
    Final_dataframe.loc[Indices, 'Principal Component-2'],               
    Final_dataframe.loc[Indices, 'Principal Component-3'],
    c = color,
    s = 70)

ax.legend(Species_types)
ax.grid()
plt.show()
```
